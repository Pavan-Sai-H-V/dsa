# filehandling

- **'r'**: Read (default mode). Opens the file for reading.
- **'w'**: Write. Opens the file for writing (creates a new file or truncates an existing file).
- **'a'**: Append. Opens the file for appending (creates a new file if it doesn't exist).
- **'x'**: Exclusive creation. Creates a new file and fails if the file already exists.
- **'b'**: Binary mode. Used with other modes (e.g., 'rb', 'wb') for binary files.
- **'t'**: Text mode (default). Used with other modes (e.g., 'rt', 'wt') for text files.
- **'+'**: Update mode. Opens the file for both reading and writing (e.g., 'r+', 'w+').
- **'r+'**: Read + Write
- **'w+'**: Write + Read
- **'a+'**: append + Read
- **'rb'**: read binary
- **'wb'**: Write binary
- **'ab'**: append binary

## writing 'w'

```python
file = open("notes.txt",'w')
file.write("welcome to py file handling\n")
file.write('This is a new file \n')
file.close()
```

## reading 'r' 
overwrites completely

```python
file_r = open("notes.txt","r")
content = file_r.read()
print("file Content\n",content)
file.close() 
```

## appending 'a'
### adds without rewriting

```python
file=open('notes.txt','a')
file.write('We added a new line using append :)')
file.close()
```

```python
with open("notes.txt",'r') as newfile:
    for l in newfile:
        print(l)
```

```python
with open('data.txt','r') as f:
    print(f.readline())
```

```python
with open('data.txt','r') as f:
    for _ in range(10):
        print(f.readline().strip())
        
```

```python
with open("data.csv",'r') as file, open('replica.csv','w') as copy:
    for l in file:
        print(l.strip())
        copy.write(l)
        
```
