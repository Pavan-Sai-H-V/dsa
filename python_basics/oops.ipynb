# Access Specifiers

```python
class Parent:
    def normal_method(self):
        print('Im a normal method')

    def _protect_method(self):
        print('Im a protected  method')

    def __private_method(self):
        print('Im a private  method')

    def acess_inside_the_parent_classs(self):
        print('Inside parent class')
        self.normal_method()
        self._protect_method()
        self.__private_method()

    
p=Parent()
p.acess_inside_the_parent_classs()
```

```python
class Child(Parent):
    def acess_inside_the_Child_classs(self):
        print('Access from child class')
        self.normal_method()
        self._protect_method()
        try:
            self.__private_method()
        except AttributeError:
            print("Not possible")


c=Child()
c.acess_inside_the_Child_classs()
```

```python
class Stranger():
    def acess_inside_the_Child_classs(self):
        print('Access from child class')
        self.normal_method()
        self._protect_method()
        try:
            self.__private_method()
        except AttributeError:
            print("Not possible")


c=Child()
c.acess_inside_the_Child_classs()
```

# ** ENCAPSULATION**

```python
class Order :
    def __init__(self,customer_name,items,tot_amt,discount):
        self.customer_name=customer_name
        self.items=items
        self.__tot_amt=tot_amt
        self.__discount=discount

    def __calculate(self):
        return self.__tot_amt - self.__discount
    

    def _get_admin_view(self):
        return{
            "Customer":self.customer_name,
            "Items":self.items,
            "Total Amount": f"${self.__tot_amt}",
            "Discount Appllied":f"${self.__discount}",
            "Final Bill": f"&{self.__calculate()}"
        }
    
    def get_customer_view(self):
        return {
            "Customer": self.customer_name,
            "Items": self.items,
            "Final Bill": f"${self.__calculate()}"
        }
    

class AdminPortal:
    def showOrder(self,order):
        return order._get_admin_view()
    
class CustomerApp:
    def show_order(self,order):
        return order.get_customer_view()
    

order1 = Order("John Doe", ["Book", "Pen"], 50, 5)
admin = AdminPortal()
customer = CustomerApp()
print('Admin view')
print(admin.showOrder(order1))
print('Customer view')
print(customer.show_order(order1))
```

# Static Method

```python
class Math:
    @staticmethod

    def add(a,b):
        return a+b
    
print(Math.add(2,4))
```

```python
class  Employee:
    company='OpenAI'
    @classmethod

    def change_company(cls,new_name):
        cls.company = new_name

    @staticmethod
    def try_change(new_name):
       company=new_name

Employee.change_company("Google")
print("After classmethod",Employee.company)

Employee.try_change("meta")
print("After static method",Employee.company)
```
